# Linear Regression

**Using user defined Functions**

```python
# functions 
# calculating the values of b0 and b1
def coeff(x,y):
    n = len(x)
    
    mean_x = sum(x)/n
    mean_y = sum(y)/n
    
    SS_xy = sum(x*y) - n*mean_x*mean_y
    SS_xx = sum(x*x) - n*mean_x*mean_x
    
    b_1 = SS_xy/SS_xx
    b_0 = mean_y -mean_x*b_1
    
    return (b_0,b_1)
```

```python
# function to plot the regression line 
def plotregressionline(x,y,b):
    plt.scatter(x,y,color = 'r')
    
    y_pred = b[0] + b[1]*x
    
    plt.plot(x,y_pred,color= 'b')
    
    plt.title('Simple linear regression using user defined functions')
    plt.xlabel('x')
    plt.ylabel('y')
    
    plt.show()
```

```python
def coeff_of_det(x, y):
  n = len(x)
  r2_num = (n*sum(x*y) - sum(x)*sum(y))**2
  r2_den = (n*sum(x*x) - (sum(x))**2) * (n*sum(y*y) - (sum(y))**2)
  r_2 = r2_num/r2_den
  return r_2
```

# VIF (Variance Inflation Factor)

```python

def VIF(x,y):
    r = coeff_of_det(x,y)
    VIF = 1/(1-r)
    return VIF
```

```python
def predicted(x,b0,b1):
    y_pred = b_0 + b_1*x
    return y_pred
```

```python
def SSR(y_mean , y_pred):
    SSr_value = sum((y_mean - y_pred)**2)
```

```python
def SST(y_mean,y_i):
    SST_value = sum((y_mean - y_i)**2)
    return SST_value 
```

```python
def SSE(y_pred,y_i):
    SSE_value = sum((y_pred - y_i)**2)
    return SSE_value
```

```python
def slopeandintercepts(x1,x2,y):
    sumofx2square = sum(x2*x2) - (sum(x2)*sum(x2))/len(x2)
    sumofx1square = sum(x1*x1) - (sum(x1)*sum(x1))/len(x1)
    sumofx1y = sum(x1*y) - (sum(x1*y))/len(x1)
    sumofx2y = sum(x2*y) - (sum(x2*y))/len(x2)
    sumofx1x2 = sum(x1*x2) - (sum(x1)*sum(x2))/len(x2)
    
    # slopes of x1 and x2 respectively
    b1 = (((sumofx2square)*(sumofx1y)) - ((sumofx1x2)*(sumofx1y)))/(((sumofx1square)*(sumofx2square)) - (sumofx1x2)*(sumofx1x2))
    b2 = (((sumofx1square)*(sumofx2y)) - ((sumofx1x2)*(sumofx1y)))/(((sumofx1square)*(sumofx2square)) - (sumofx1x2)*(sumofx1x2))
    
    # mean of y,x1 and x2
    y_mean = sum(y)/len(y)
    x1_mean = sum(x1)/len(x1)
    x2_mean = sum(x2)/len(x2)
    
    # calculating the slope 
    b0 = y_mean - b1*x1_mean - b2*x2_mean
    
    return b0,b1,b2
    
```

# Confusion Matrix 

```python
# function to compute the confusion matrix 

def confusionmatrix(actual,predicted):
    
    # the classes during making of confusion matrix should be unique 
    classes = np.unique(actual)
    
    # creating a null matrix 
    confusion_matrix = np.zeros((len(classes),len(classes)))
    
    # looping around the actual to predicted with nested loop to cover each combination
    for i in range(len(classes)):
        for j in range(len(classes)):
            
            confusion_matrix[i,j] = np.sum((actual == classes[i]) & (predicted == classes[j]))
            
    return confusion_matrix 
```

```python
actual = [1, 3, 3, 2, 5, 5, 3, 2, 1, 4, 3, 2, 1, 1, 2]
predicted = [1, 2, 3, 4, 2, 3, 3, 2, 1, 2, 3, 1, 5, 1, 1]
```

```python
matrix = confusionmatrix(actual,predicted)
matrix
```

```python
total_sum = 0
tot_sum = 0
for i in range(0, len(matrix)):
  for j in range(0, len(matrix)): 
    tot_sum += matrix[i][j]
tot_sum
```

```python
# Hence we've a case of multi class classification
def params(mat, n):
  col_sum = [sum([row[i] for row in mat]) for i in range(0,len(mat[0]))]
  row_sum = [sum(mat[i]) for i in range(len(mat))]
  tp = mat[n][n]
  fp = row_sum[n] - tp
  fn = col_sum[n] - tp
  tn = tot_sum - fp - fn - tp
  param = []
  accuracy = (tp + tn) / (tp + tn + fp + fn)
  param.append(accuracy)
  precision = tp / (tp + fp)
  param.append(precision)
  recall = tp / (tp + fn)
  param.append(recall)
  f1_score = (2 * precision * recall) / (precision + recall)
  param.append(f1_score)
  return param
```

```python
for i in range(len(matrix)):
  print(f"For the class {i+1}, the values are as follows: ")
  print(f"Accuracy: {params(matrix,i)[0]}")
  print(f"Precision: {params(matrix,i)[1]}")
  print(f"Recall: {params(matrix,i)[2]}")
  print(f"F1-Score: {params(matrix,i)[3]}")
  print("*********************")
```
